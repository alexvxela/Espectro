## TRANSMITANCIA DE LOS FILTROS.

```
# Conversión del espectro sin el filtro a escala de grises
I0 = Image.open("01-PreFiltros.JPG")
plt.subplot(1,2,1)
plt.imshow(np.asarray(I0))
plt.title('Luz Blanca')

I1 = I0.convert('L')
c1 = np.asarray(I1, dtype=np.float32)
plt.subplot(1,2,2)
plt.imshow(c1, cmap='gray')
plt.title('Gris')
plt.axis('Off')
```

**Absorbancia del beta-caroteno.**

Suponiendo que no se refleja la luz a través del líquido (jugo de zanahoría filtrado y diluído en agua), entonces la luz que éste absorbe corresponde a la diferencia entre la luz que se transmite a través del recipiente y la que se transmite a través de ambos (líquido y recipiente).

```
# Graficas: difraccion recipiente, líquido / Absorbancia
y7 = y5 - y6

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12, 5))
fig.suptitle('Perfiles de la fuente, el recipiente y el líquido. Absorbancia beta-caroteno', fontsize = 15)

ax1.plot(w, y5/234., color='blue', label='Recipiente'), ax2.plot(w, y7/234., 'tab:blue')
ax1.plot(w, y6/234., color='orange', label='Liquido')
ax1.plot(w, y4/234., color='purple', label='Perfil Fuente')
ax1.set_xlabel('Longitud de Onda λ [nm]', fontsize = 12)
ax1.set_ylabel('Intensidad Normalizada [%]', fontsize = 12) # Unidades Arbitrarias
ax2.set_xlabel('Longitud de Onda λ [nm]', fontsize = 12)
ax2.set_ylabel('Absorbancia [%]', fontsize = 12)
ax1.legend(), #ax2.legend()
ax1.grid(), ax2.grid()
plt.subplots_adjust(wspace=0.3)
plt.show()
```

```
# Transmitancia

t = 1 - y7/234.

plt.figure(figsize=(6,4.5))
plt.title('Transmitancia del beta-caroteno', fontsize = 15)
plt.plot(w, t/1.5, color='orangered')
plt.xlabel('Longitud de Onda λ [nm]', fontsize = 12)
plt.ylabel('Transmitancia, [%]', fontsize = 12)
#plt.axvline(x = E[2], linestyle='--', color='r', label="800 nm")
#plt.axvline(x = E[1], linestyle='--', color='orange', label="600 nm")
#plt.axvline(x = E[0], linestyle='--', color='b', label="400 nm")
plt.grid()
plt.show()
```

```
# Celda para obtener máximos locales
x_min = 450
x_max = 600

rango_x = l[(l > x_min) & (l < x_max)]
rango_y = list(y_v[(l > x_min) & (l < x_max)])
xmax = rango_x[rango_y.index(max(rango_y))]

print("El maximo local esta en", xmax)
print("Su intensidad es",max(rango_y))
```
